## Hypothesis (1変数の線形回帰)

```python
def Hypothesis(theta,X):

    #h(theta) = theta0 * 1 + theta1 * x
    #行列の積はdot,要素同士の掛け算は * で表現する.
    return np.dot(X,theta)  # 行列の積/Sumと同義
```

## CostFunction(最急降下法を用いてこの値を最小にする)

```python
def CostFunction(theta,X,Y,m,i):
    
    Prediction = 0
    Cost = 0
    
    Prediction =  (Hypothesis(theta,X) - Y )    
    Cost = (1 / 2 * m) * np.sum(Prediction * Prediction) #numpyの2乗は ** で実装!!!

    if i % 10 == 0 :
        print("与えられたTheta:" + str(theta))
        print("予測結果:         "+ str(np.dot(X,theta)))
        print("予測との誤差:   " + str(Prediction))    
        print("計算したコスト:  " + str(Cost)) 
    
    return Cost
```

## Gradient Descent (更新処理)による最小化問題

```python
def GradientDescent(theta, X,  y, m , num_iter = 100):
    alpha = 0.01  # Learning Rate
    ygraph = [] #コスト関数の返す値を可視化するためのリスト

    #初期値による予測とコスト計算
    print("---- " + "初期値でのコスト計算 ----")
    ygraph.append(CostFunction(theta,X,y,m,0)) 
        
    #Gradient Descentの実行 
    for i in range(num_iter):
    
        if i % 10 == 0 :
            print("---- " + str(i+1) + "回目の更新作業 ----")
        
        
        h = Hypothesis(theta,X)
        
        print("theta:" + str(theta.shape))
        print("h-y:" + str((h-y).shape))
        print("X:"+ str(X.shape))
        print("np.dot(h-y).T,X :" +str((np.dot((h - y).T,X).T.shape)))
        
        theta =  theta - alpha  /  m  * (np.dot((h - y).T,X)).T
        ygraph.append(CostFunction(theta,X,y,m,i))
                          
    #アルゴリズム実行後の結果表示
    print("----Gradient Descent 終了----")
    print("最終パラメータ予測結果" + str(theta))

    
    #コスト関数の可視化処理
    
    #x軸に表示するアルゴリズムの実行回数
    xgraph = [x+1 for x in range(num_iter+1)]
    plt.plot(xgraph,ygraph)
    plt.xlabel("times of performing Gradient Descent")
    plt.ylabel("value of computing cost function")
 
```
